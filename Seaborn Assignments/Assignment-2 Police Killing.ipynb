## 1b: Swarm Plot

## 2: Plot 12 most dangerous cities

```python
df.city.value_counts().head(12)
```

```python
# with seaborn barplot

```

```python
plt.figure(figsize=(8,5))

sns.barplot(data=df, x=df.city.value_counts().head(12).index, y=df.city.value_counts().head(12))

plt.title("Most dangerous cities", color="b", fontsize=15)
plt.xticks(rotation=45)

plt.show()
```

```python
# with seaborn countplot
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x="city", order=df.city.value_counts().head(12).index)

plt.title("Most dangerous cities", color="b", fontsize=15)
plt.xticks(rotation=45)

plt.show()
```
